# **プログラムの名称**

## **概要**

じゃんけんができるプログラムです。

普通のじゃんけんの他にも、

・五本先取勝負（先に五勝した方が勝利です）

・十回勝負（十回中何回勝てるかを競います）

・あっちむいてホイ

を実装しています。



## **操作方法**

全て数字で操作します。数字は全角でも半角でも大丈夫です。

操作方法は下の通りです。

◆じゃんけんの時に、自分の出す手を決定する方法

　　1：グー、2：チョキ、3：パー　で決定します。

◆あっちむいてホイの時に、自分が向く方向（もしくは、指をさす方向）を決定する方法

　　1：↑、2：↓、3：←、4：→　で決定します。

```python
#@title メイン処理、ユーザー入力部分

#勝利回数、敗北回数を計測します。
counter_v = 0
counter_d = 0

#遊ぶモードを設定する場所です
print("1～4までの数字を入力し、遊びたいモードを選んでください。")
print("　1:ノーマルモード")
print("　2:五本先取")
print("　3:十回勝負")
print("　4:あっちむいてホイ")

#全角で入力された数字を半角に変換する処理です。
mode = unicodedata.normalize('NFKC',input())

#何も入力されなかった時に入力し直させる処理です。
while mode == "":
  print("モードの設定は必須です。もう一度入力し直してください。")
  mode = input()

#指定した値以外が入力された時に入力し直させる処理です。
searchWord_mode = "1 2 3 4"
while mode not in searchWord_mode:
  print("無効な値が入力されました。もう一度入力し直してください。")
  mode = input()

#入力された文字列を数値に変換します。
mode = int(mode)

print("ルール：", end=(''))
if mode == 1:
  print("普通のじゃんけんです。勝つか負けるまで続きます。")
elif mode == 2:
  print("先に5勝した方の勝利です。あいこは勝利数にカウントしません。")
elif mode == 3:
  print("10回中何回勝てるかを競います。あいこは勝利数にカウントしません。")
else:
  print("普通のあっちむいてホイです。長期戦になりがちなので、頑張って下さいね。")

print("-------------------------------------------------------")

#五本先取、十回勝負のの時に繰り返すための処理です。
for i in range(100):
  #プレイヤーが出す手を決定する場所です
  print("1～3までの数字を入力し、自分の出す手を決めてください。（1:グー、2:チョキ、3:パー）")
  player = unicodedata.normalize('NFKC',input())

  #何も入力されなかった時に入力し直させる処理です。
  while player == "":
    print("自分の出す手が指定されませんでした。もう一度入力し直してください。")
    player = input()

  #指定した値以外が入力された時に入力し直させる処理です。
  searchWord_player = "1 2 3"
  while player not in searchWord_player:
    print("無効な値が入力されました。もう一度入力し直してください。")
    player = input()

  #入力された文字列を数値に変換します。
  player = int(player)

  #今が何回戦目かを分かりやすくします。
  #ただし1回戦で終わるモードでは表示しません。
  if mode == 2 or mode == 3:
    print("-------------------------------------------------------")
    print(f"第{i+1}回戦")

  #プレイヤーの入力した数値を手の絵文字に変換します
  you = hand_list[player-1]
  print("あなた：{}".format(you))

  #相手の出す手を決定します
  enemy = r.randint(0,2)
  com = hand_list[enemy]
  print("相手：{}".format(com))

  #じゃんけんの勝敗を判定します。
  #あいこのときの処理です。
  if you == com:
    print("あいこです。")
  #あなたが勝った事を判断する処理です。
  elif player == enemy or player-3 == enemy:

    #あっちむいてホイの時のみ処理されます。
    if mode == 4:
      #返された値を受け取る場所です。
      v = arrow_v()
      #勝った時に勝負を終わらせる処理です。
      if v == victory:
        victory()
        counter_v = 1

    #あなたが勝った時に、勝った回数をカウントする処理です。
    else:
      victory()
      counter_v += 1
  #あなたが負けた事を判断する処理です。
  else:

    #あっちむいてホイの時のみ処理されます。
    if mode == 4:
      #返された値を受け取る場所です。
      d = arrow_d()
      #負けた時に勝負を終わらせる処理です。
      if d == defeat:
        defeat()

    #あなたが負けた時に、負けた回数をカウントする処理です。
    else:
      defeat()
      counter_d += 1

  print("-------------------------------------------------------")

  #ノーマルモード、もしくは、あっちむいてホイの時に、勝利したら終わらせる為の処理です。
  if mode == 1 or mode == 4:
    if counter_v == 1 or counter_d == 1:
      break

  #五本先取の時に、あなたと相手が何勝しているか表示させる処理です。
  if mode == 2:
    print(f"あなた：{counter_v}勝")

    #あなたが4勝したら知らせてくれる処理です。
    if counter_v == 4 and counter_d < 4:
      print("☆★あと1勝で勝利です！★☆")
    print(f"相手：{counter_d}勝")

    #あなたも相手が4勝したら知らせてくれる処理です。
    if counter_d == 4 and counter_v == 4:
      print("□■お互いにあと1勝です！！■□")

    #相手が4勝したら知らせてくれる処理です。
    elif counter_d == 4 and counter_v < 4:
      print("◇◆相手がリーチです！◆◇")

    #5勝または5敗したら終わらせる処理です。
    if counter_v == 5:
      print("☆★先に5勝したので、あなたの勝ちです！！★☆")
      break
    elif counter_d == 5:
      print("◇◆先に相手に5勝されたので、あなたの負けです...◆◇")
      break

  #十回勝負の時に、十回で終わらせる為の処理です。
  if mode == 3:
    if i == 9:
      if counter_v == 0:
        print('1回も勝てませんでした...')
      else:
        print(f'●〇10回中{counter_v}回勝利しました！〇● ')
      break
```
